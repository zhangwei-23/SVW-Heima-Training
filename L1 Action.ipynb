# Action2:统计全班的成绩

```python
#coding utf-8
import numpy as np
import pandas as pd
```

```python
#numpy
persontype = np.dtype({
   'names':['name','chineses','math','English'],
   'formats':['S32','i','i','i']})

peoples = np.array(
    [   ("zhangfei", 68, 65, 30),
        ("guanyu", 95, 76, 98),
        ("liubei", 98, 86, 88),
        ("dianwei", 90, 88, 77),
        ("xuchu", 80, 90, 90)
    ], dtype=persontype)


chineses = peoples[:]['chineses']
math = peoples[:]['math']
English = peoples[:]['English']

def show(name,cj):
    print('{} | {} | {} | {} | {} | {} '
          .format(name,np.mean(cj),np.min(cj),np.max(cj),np.var(cj),np.std(cj)))
    
print("科目 | 平均成绩 | 最小成绩 | 最大成绩 | 方差 | 标准差")
show("语文", chineses)
show("英语", English)
show("数学", math)
```

```python
ranking = sorted(peoples,key=lambda x:x[1]+x[2]+x[3], reverse=True)
print( ranking) #打印排名
```

# Action3:对汽车质量数据进行统计

```python
data = pd.read_csv(r'D:\1_Up\Heima\Data_Engine\L1\car_data_analyze\car_complain.csv')
```

```python
#data.shape
data.head()
```

```python
data = data.drop('problem',1).join(data.problem.str.get_dummies(','))
data.head()
```

```python
#品牌投诉总数

def f(x):
    x = x.replace('一汽-大众','一汽大众')     #数据清洗，合并同名
    return x

data['brand'] = data['brand'].apply(f)    
result1 = data.groupby(['brand'])['id'].agg(['count'])
result1
```

### 不同problem类型的总数

```python
tags = data.columns[7:]
tags
```

```python
result2 = data.groupby(['brand'])[tags].agg(['sum'])
result2
```

### 品牌投诉总数排序

```python
result3 = result1.merge(result2,left_index=True,right_index=True,how = 'left')
result3.reset_index(inplace=True)
result3 = result3.sort_values('count',ascending =False)
result3
```

### 车型投诉总数排序

```python
result4 = data.groupby(['car_model'])['id'].agg(['count'])
result4 = result4.sort_values('count',ascending =False)
result4.reset_index(inplace=True)
result4
```

### 品牌的车型平均投诉 

```python
result5 = data.groupby(['brand','car_model'])['id'].agg(['count'])
result5
```
